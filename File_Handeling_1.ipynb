```python
# reading from files using read()
f = open("sample3.txt",'r')
s = f.read()
print(s)
f.close()
```

```python
# reading upto n character
f = open("sample3.txt",'r')
s = f.read(18)
print(s)
f.close()
```

```python
# read line by line
f = open("sample1.txt",'r')
print(f.readline(),end='')
print(f.readline(),end='')
f.close()
```

```python
# reading entire using readline
f = open("sample.txt",'r')
while True:
    data = f.readline()
    
    if data == '':
        break
    else:
        print(data,end='')
f.close()
```
