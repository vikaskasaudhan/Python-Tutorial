```python
class Fraction:

    def __init__(self,x,y):
        self.num = x
        self.den = y

    def __str__(self):
        return '{}/{}'.format(self.num,self.den)

    def __add__(self,other):
        new_num = self.num*other.den + self.den*other.num
        new_den = self.den*other.den
        return '{}/{}'.format(new_num,new_den)

    def __sub__(self,other):
        new_num = self.num*other.den - self.den*other.num
        new_den = self.den*other.den
        return '{}/{}'.format(new_num,new_den)


    def __mul__(self,other):
        new_num = self.num*other.num 
        new_den = self.den*other.den
        return '{}/{}'.format(new_num,new_den)

    def __truediv__(self,other):
        new_num = self.num*other.den
        new_den = self.den*other.num
        return '{}/{}'.format(new_num,new_den)

    def convert_to_decimal(self):
        return self.num/self.den
```

```python
fr1 = Fraction(2,3)
fr2 = Fraction(4,3)
```

```python
print(fr1)
print(fr1+fr2)
print(fr1-fr2)
print(fr1*fr2)
print(fr1/fr2)
print(fr1.convert_to_decimal())
```

```python
class Rectangle:

    def __init__(self,l,w):
        self.length = l
        self.width = w

    def area(self):
        return self.length * self.width
    
    def perimeter(self):
        return 2*(self.length+self.width)
    
    def display(self):
        print("length of rectangle = ",self.length)
        print("width of rectangle = ",self.width)
        print("area of rectangle = ",self.area())
        print("peremeter of rectangle = ",self.perimeter())
```

```python
rec = Rectangle(2,3)
```

```python
rec.display()
```

```python
class BankAccount:
    def __init__(self,an,n,b):
        self.account_num = an
        self.name = n
        self.balance = b

    def deposite(self,bal):
        self.balance += bal
    
    def withdrawl(self,amount):
        self.balance -= amount

    def bankFees(self,year):
        self.balance += 0.05*self.balance*year

    def display(self):
        print("Account number = ",self.account_num)
        print("Account name = ",self.name)
        print("Account balance = ",self.balance)


newAccount = BankAccount(42270716363,'Vikas Kasaudhan',10000)
newAccount.deposite(5000)
newAccount.bankFees(3)
newAccount.withdrawl(5000)
newAccount.display()
```

```python
class Computation:

    def __init__(self):
        pass
    
    def factorial(self):
        num = int(input("enter positive integer to calculate factorial"))
        fact = 1
        for i in range(1,num+1):
            fact *= i
        print("factorial of ",num, "=",fact)

    def naturalSum(self):
        num = int(input("enter positive integer"))
        sum = 0
        for i in range(1,num+1):
            sum += i
        print("sum of first ",num,"numbers = ",sum)

    def testPrime(self,num):
        flag = True
        for i in range(2,num//2+1):
            if num%i==0:
                flag = False
                break
        if flag:
            return "Prime"
        else:
            return "Not Prime"

    def testPrims(self,a,b):
        common_div = 1
        for i in range(1,min(a,b)+1):
            if a%i==0 and b%i==0:
                common_div = i
        if common_div==1:
            return "Yes"
        else:
            return "No"
        
    def tableMult(self):
        num = int(input("enter integer to to display multplication table"))
        print(num,":",end=" ")
        for i in range(1,11):
            print(num*i,end=" ")

    def allTableMult(self):
        for i in range(1,10):
            print(i,":",end=" ")
            for j in range(1,11):
                print(i*j,end=" ")
            print("\n")

    def listDiv(self,num):
        Ldiv = []
        for i in range(1,num+1):
            if num%i==0:
                Ldiv.append(i)
        return Ldiv
    
    def listPrimeDiv(self,num):
        Ldiv = []
        for i in range(2,num+1):
            if num%i==0:
                flag = True
                for j in range(2,i//2+1):
                    if i%j==0:
                        flag = False
                        break
                if flag:
                    Ldiv.append(i)
        return Ldiv
```
