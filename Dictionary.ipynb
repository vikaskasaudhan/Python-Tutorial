Dictionary in Python is a collection of keys values, used to store data like a map , which unlike other data types which hold only a single value as an element

in some languages it is known as map or associative arrays

Characteristics
* Mutable
* Indexing has no meaning
* Keys can not be duplicated
* keys can not be mutable items

## Creating

```python
# empty
d = {}
print(d,type(d))

# 1D dict
d = {'name':'vikas','age':21,'degree':'B Tech'}
print(d)

# with mixed keys
d = {(1,2,3):1,'hello':'world'}
print(d)

# 2D dict
d = {'name':'vikas','age':21,'college':'NIT DGP','semester':3,
    'subjects':{
        'math':99,
        'dsa':78,
        'oop':89
    }
    }
print(d)
```

```python
# type conversion
d = dict([('name','vikas'),('age',21),('marks',87)])
print(d)

# duplicates
d = {'name':'vikas','name':'kasaudhan'}
print(d)
```

```python
# mutable items as keys
d = {'name':'vikas',[1,2,3]:3}
print(d)
```

```python
d = {'name':'vikas',(1,2,3):3}
print(d)
```

## Remove key-value pair

```python
dic = {'name':'vikas','age':21,'gender':'male',3:9}
# pop
dic.pop(3)
print(dic)

# popitem
dic.popitem()
print(dic)

#del
del dic['age']
print(dic)

# clear
dic.clear()
print(dic)

#del
del dic
print(dic)
```

## Editing key-value pair

```python
dic = {'name':'vikas','age':21,'gender':'male',3:9}
dic['name'] = "Madhuri"
dic['gender'] = 'female'
print(dic)
```

## Dictionary Operations

* Membership (works on key only)
* Iteration

```python
dic = {'name':'vikas','age':21,'gender':'male',3:9}
'name' in dic
```

```python
dic = {'name':'vikas','age':21,'gender':'male',3:9}

for i in dic:
    print(i, dic[i])
```

## Dictionary Functions

```python
# len / sorted
dic = {'name':'vikas','age':21,'gender':'male',3:9}
print(len(dic))
sorted(dic,key=str)
```

```python
# items / keys / values
dic = {'name':'vikas','age':21,'gender':'male',3:9}
print(dic)
print(dic.items())
print(dic.keys())
print(dic.values())
```

```python
# update
d1 = {1:2,3:4,4:5}
d2 = {4:7,6:8}
d1.update(d2)
print(d1)
```

## Dictionary Comprehension

{ key : value for vars in iterable }

```python
# print 1st 10 numbers and their squires
{i:i**2 for i in range(1,11)}
```

```python
# using existing dictionary
distances = {'delhi':1000,'mumbai':1700,'bangalore':2500}
{key:value*0.62 for (key,value) in distances.items()}
```

```python
# Using zip
days = ['Sunday','Monday',"Tuesday","Wednesday","Thursday","Friday","Saturday"]
temp_C = [30.5,39.7,45,43.9,29.9,34,28]

{i:j for (i,j) in zip(days,temp_C)}
```

```python
# Using if Condition
products = {'phone':10,'laptop':0,'charger':32,'tablet':0}

{key:value for (key,value) in products.items() if value>0}
```

```python
# Nested Comprehension
# print tables of number from 2 to 5

{i:{j:i*j for j in range(1,11)} for i in range(2,5)}
```
