```python
import nlpcloud

class NLPapp:
    def __init__(self):
        self.__database = {}
        self.__first_menu()

    def __first_menu(self):
        first_input = input("""
        Hello ! How would you like to proceed
        1. Not a member? Register.
        2. Already a member? Login.
        3. Galti se aa gaya? Exit.
        """)

        if first_input=='1':
            self.__register()
        elif first_input=='2':
            self.__login()
        else:
            exit()


    def __second_menu(self):
        second_input = input("""
        Hello ! How would you like to proceed
        1. NER.
        2. Language Detection.
        3. Sentiment.
        4. Chatbot.
        5. codegenerate.
        6. grammer and spelling correction.
        7. Logout.
        """)

        if second_input=='1':
            self.__ner()
        elif second_input=='2':
            self.__language_detect()
        elif second_input=='3':
            self.__sentiment()
        elif second_input=='4':
            self.__chatbot()
        elif second_input =='5':
            self.__codegeneration()
        elif second_input=='6':
            self.__grammer_spelling_correction()
        else:
            exit()

    def __register(self):
        name = input("enter name")
        email = input("enter email")
        password = input("enter password")
        if email in self.__database:
            print("email already exist")
        else:
            self.__database[email] = [name,password]
            print("registration successful. Now Login")
            print(self.__database)
            self.__login()

    def __login(self):
        email = input("enter email")
        password = input("enter password")
        if email in self.__database:
            if self.__database[email][1]==password:
                print("login successful")
                self.__second_menu()
            else:
                print("wrong password")
                self.__login()
        else:
            print("this email is not registered")
            self.__first_menu()
    

    def __ner(self):
        para = input("enter paragraph")
        search_term = input("what would you like to search")

        client = nlpcloud.Client("finetuned-llama-3-70b", "56c5247ee2b99bde57bd5a5d9ab4ee39cb72de3c", gpu=True)
        response = client.entities(para , searched_entity= search_term)
        print(response)

    def __language_detect(self):
        para = input("enter paragraph")

        client = nlpcloud.Client("python-langdetect", "56c5247ee2b99bde57bd5a5d9ab4ee39cb72de3c", gpu=False)
        response = client.langdetection(para)
        print(response)

    def __sentiment(self):
        para = input("enter paragraph")
        client = nlpcloud.Client("finetuned-llama-3-70b", "56c5247ee2b99bde57bd5a5d9ab4ee39cb72de3c", gpu=True)
        response = client.sentiment(para , target="NLP Cloud")
        print(response)

    def __chatbot(self):
        para = input("enter your sentence")
        client = nlpcloud.Client("finetuned-llama-3-70b", "319bd9d66fe4efcb62d41b48f26fbe7b8a75f52a", gpu=True)
        responce = client.chatbot(para, context="""This is a discussion between a human and an AI. The human is sad but the AI is empathetic and reassuring. The AI is called Patrick.""",
                    history=[]
                    )
        print(responce)
        num = input("enter 1 to talk again")
        if num == '1':
            self.__chatbot()
        else:
            exit()


    def __codegeneration(self):
        para = input("enter sentence to generate code")
        client = nlpcloud.Client("finetuned-llama-3-70b", "7f44936d881a16852b26f26376af9f6f8fa32a70", gpu=True)
        response = client.code_generation(para)
        print(response)


    def __grammer_spelling_correction(self):
        para = input("enter sentence to correct grammer and spelling")
        client = nlpcloud.Client("finetuned-llama-3-70b", "7f44936d881a16852b26f26376af9f6f8fa32a70", gpu=True)
        response = client.gs_correction(para)
        print(response)
        
```
