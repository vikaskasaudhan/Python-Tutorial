## Fruit Quiz

```python
import numpy as np
import pandas as pd
import random
```

```python
class FlashCard:

    def __init__(self):
        pass

    def init(self):
        print("welcome to fruit quiz:")
        while True:
            dic = {"banana":"yellow","apple":"red","grapes":"green","orange":"orange","strawberry":"pink"}
            key = random.choice(list(dic.keys()))
            val = dic[key]
            print("what is the color of the ",key)
            user_input = input("")
            if user_input==val:
                print("correct answer")
            else:
                print("wrong answer try again")
            user_choice = input("enter 0 to play again :")
            if type(user_choice)==int:
                if user_input==0:
                    continue
            else:
                start_again = input("enter start to start again")
                if type(start_again)==str:
                    if start_again=="start":
                        self.init()
                else:
                    print("you entered wrong choice type or wrong choice")
                    break
```
