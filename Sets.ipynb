# Sets

A set is an unordered collection of items. Every set element is unique (no duplicates) and must be immutable (can not be changed).
However, a set itself is mutable. We can add or remove items from it.
Sets can also mbe used to perform mathematical set operation like union, intersection, symmetric difference etc.

Characteristics
* Unordered
* Mutable
* No Duplicates
* Can not contain mutable data type

## Creating

```python
# empty
s = {}
print(s)
print(type(s))
s = set()
print(s)
print(type(s))

# 1D and 2D
s = {1,2,3,4}
print(s)
s = {1,2,3,{5,6,8}}
print(s)
```

```python
s = {1,2,3,4,True,'vikas',(6,7,8)}
print(s)
# not printing True because duplicate
# using type conversion
s = set([1,2,3,4])
print(s)
```

```python
# set can not have mutable items
s = {1,2,3,4,[6,7,8]}
print(s)
```

```python
s1 = {1,2,3}
s2 = {3,2,1}
print(s1==s2)
# because of unordered
```

## Accessing

```python
# can not indexing and slicing
s = {1,2,3,4}
print(s[0])
```

## Editing

```python
# can not edit
s = {1,2,3,4}
s[0] = 100
print(s)
```

## Adding items

```python
s = {1,2,3,4,5}
# add
s.add(8)
print(s)

# update
s.update([6,7,8,9])
print(s)
```

## Deleting

```python
# del
s = {1,2,3,4}
print(s)
del s
print(s)
```

```python
# discard
s = {1,2,3,4,5}
s.discard(5)
print(s)
s.discard(8)
print(s)
```

```python
# remove
s = {1,2,3,4,5,6,7,8,89}
s.remove(5)
print(s)
s.remove(20)
print(s)
```

```python
# pop
s = {2,3,4,5,6,7,8,9}
s.pop()
```

```python
# clear
s = {2,3,4,5,6,7,8,9}
s.clear()
print(s)
```

## set operation

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8,9}
# Union(|)
print(s1 | s2)

# intersection
print(s1 & s2)

 # difference
print(s1-s2,s2-s1)

# symmetric difference
print(s1^s2,s2^s1)

# Membership
print(1 in s1)
print(1 not in s1)

# loop
for i in s1:
    print(i)
```

## Functions

```python
# len / sum / min / max / sorted
s = {1,2,6,3,4,8,7}
print(len(s))
print(max(s))
print(sum(s))
print(sorted(s))
```

```python
# union / update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8,9}
print(s1.union(s2))
```

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8,9}
s1.update(s2)
print(s1)
print(s2)
```

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8,9}
print(s1.intersection(s2))
```

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8,9}
s1.intersection_update(s2)
print(s1)
print(s2)
```

```python
# difference / difference_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8,9}
print(s1.difference(s2))
s1.difference_update(s2)
print(s1)
print(s2)
```

```python
# symmetric_difference / symmetric_difference_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8,9}
print(s1.symmetric_difference(s2))
s1.symmetric_difference_update(s2)
print(s1)
print(s2)
```

```python
#  isdisjoint / issubset / issuperset
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8,9}
print(s1.isdisjoint(s2))
print(s1.issubset(s2))
print(s1.issuperset(s2))
```

```python
# copy
s1 = {1,2,3,4,5}
s2 = s1.copy()
print(s1,s2)
print(id(s1),id(s2))
```

# Frozenset
Frozenset is just an immutable version of a Python set Object

```python
# create 
fs = frozenset([1,2,3,4,5])
print(fs)
fs.
```

```python
# what works and does not work
# works -> all read functions
# does not work -> write operation
```

```python
fs1 = frozenset([1,2,3,4,5])
fs2 = frozenset([4,5,6,7,8])
print(fs1 | fs2)
print(fs1 & fs2)
print(fs1 ^ fs2)
print(fs1 - fs2)
```

```python
# when to use --> when we want not to change
# 2D frozenset
fs = frozenset([1,2,3,4,frozenset([5,6,7])])
print(fs)
```
