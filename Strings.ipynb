# Creating Strings

```python
s = 'vikas'
s = "vikas"
s = str('vikas')
# For multiline strings
s = '''vikas'''
s = """vikas"""
print(s)
```

# Accessing substring from a Strings

```python
# Positive Indexing
s = "Vikas Kasaudhan"
print(s[1])
```

```python
# Negative Indexing
s = "Vikas Kasaudhan"
print(s[-1])
```

```python
# Slicing String
s = "Vikas Kasaudhan"
print(s[:])
print(s[0:])
print(s[3:])
print(s[3:7])
print(s[::2])
print(s[6:0:-1])
print(s[::-1])
```

```python
s = "vikas kasaudhan"
print(s[-10:])
print(s[-10::1])
print(s[-1:-10:-1])
```

```python
# write a program to convert a string to title case without using the title()
s = input("enter string")
L = []
for i in s.split():
    L.append(i[0].upper()+i[1:])
print(" ".join(L))
```

```python
# write a program that convert an integer to string
num = int(input("enter number"))
digits = '0123456789'
res  = ''
while num>0:
    res = digits[num%10]+res
    num //= 10
print(res,type(res))
```

```python
x = int(input("enter value of x"))
n = int(input("enter number of terms n"))
sum = 0
s =''
for i in range(1,n+1):
    sum += (1/i)*(((x-1)/x)**i)
    s += '(1/{})*((x-1)/x)^{} + '.format(i,i)
print(sum)
print(s)
```
