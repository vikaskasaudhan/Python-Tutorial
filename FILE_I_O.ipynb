### Formated String Literals

* begins f or F.
* written in single or triple single or double or triple double quotation.
* Python expression written in { python expression }.

https://docs.python.org/3/tutorial/inputoutput.html#tut-f-strings

```python
name = "Vikas Kasaudhan"
age = 21
f'My name is {name} and i am {age} years old.'
```

### str.format()

https://docs.python.org/3/library/stdtypes.html#str.format

```python
name = "Vikas Kasaudhan"
age = 21
'My name is {} and i am {} years old.'.format(name,age)
```

```python
name = "Vikas Kasaudhan"
age = 21
'''My name is {} and i am {} years old.'''.format(name,age)
```

```python
name = "Vikas Kasaudhan"
age = 21
"My name is {} and i am {} years old.".format(name,age)
```

```python
name = "Vikas Kasaudhan"
age = 21
"""My name is {} and i am {} years old.""".format(name,age)
```

```python
name = "Vikas Kasaudhan"
age = 21
''''My name is {} and i am {} years old.'''''.format(name,age)
```

```python
yes_votes = 42_572_654
total_votes = 85_705_149
percentage = yes_votes / total_votes
'{:-9} YES votes  {:2.2%}'.format(yes_votes, percentage)
```

* (-) means left aligned
* 9 creates 9 spaces
* % multiply by 100 with % sign
* .2 precision after decimal
* 2 minimum total field width (smallest number of character to use) but not important because if text is longer python ignores it

```python
s = 'Hello, world.'
print(str(s))

print(repr(s))

(print)

x = 10 * 3.25
y = 200 * 200
s = 'The value of x is ' + repr(x) + ', and y is ' + repr(y) + '...'
print(s)

# The repr() of a string adds string quotes and backslashes:
hello = 'hello, world\n'
hellos = repr(hello)
print(hellos)

# The argument to repr() may be any Python object:
repr((x, y, ('spam', 'eggs')))
```

* The str() function is meant to return representations of values which are fairly human-readable.
* repr() is meant to generate representations which can be read by the interpreter  (or will force a SyntaxError if there is no equivalent syntax)
* number , lists and dictionary have the same representation with both function

```python
import math
print(f'The value of pi is approximately {math.pi:.3f}.')
```

```python
print(f'The value of pi is approximately {23.5463221:.3f}.')
```

```python
animals = 'eels'
print(f'My hovercraft is full of {animals}.')
print(f'My hovercraft is full of {animals!r}.')
```

#### used to convert the value before formatted
* '!a' applies ascii()
* '!s' applies str()
* '!r' applies repr()

```python
bugs = 'roaches'
count = 13
area = 'living room'
print(f'Debugging {bugs=} {count=} {area=}')
```

The = specifier can be used to expand an expression to the text of the expression, an equal sign, then the representation of the evaluated expression:

```python
print('{0} and {1}'.format('spam', 'eggs'))

print('{1} and {0}'.format('spam', 'eggs'))
```

```python
print('This {food} is {adjective}.'.format(
      food='spam', adjective='absolutely horrible'))
```

```python
print('The story of {0}, {1}, and {other}.'.format('Bill', 'Manfred', other='Georg'))
```

```python
for x in range(1, 11):
    print('{0:2d} {1:3d} {2:4d}'.format(x, x*x, x*x*x))
```
