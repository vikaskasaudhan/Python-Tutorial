```python
try:
    m =5
    f = open("vikas1.txt",'r')
    print(f.read())
    print(m)
    print(m/0)
except FileNotFoundError:
    print("file not found")
except NameError:
    print("variable not found")
except ZeroDivisionError:
    print("can't divide by 0")
except Exception as e:
    print(e)    
```

```python
## else block
try:
    f = open("vikas.txt",'r')
except FileNotFoundError:
    print("file not found")
except Exception:
    print("kuch to lafda h")
else:
    print(f.read())
```

```python
## finally block
try:
    f = open("vikas.txt",'r')
except FileNotFoundError:
    print("file not found")
except Exception:
    print("kuch to lafda h")
else:
    print(f.read())
finally:
    print("ye to execute hoga hi")
```

```python
class Bank:
    def __init__(self,balance):
        self.balance = balance

    def withdraw(self,amount):
        if amount<0:
            raise Exception("amount can not be -ve")
        if self.balance<amount:
            raise Exception("paise nhi h sufficient")
        self.balance = self.balance-amount

obj = Bank(10000)

try:
    obj.withdraw(15000)
except Exception as e:
    print(e)
else:
    print(obj.balance)
```

```python
class MyException(Exception):
    def __init__(self,message):
        print(message)

class Bank:
    def __init__(self,balance):
        self.balance = balance

    def withdraw(self,amount):
        if amount<0:
            raise MyException("amount can not be -ve")
        if self.balance<amount:
            raise MyException("paise nhi h sufficient")
        self.balance = self.balance-amount

obj = Bank(10000)

try:
    obj.withdraw(-15000)
except MyException as e:
    pass
else:
    print(obj.balance)
```

```python
class SecurityError(Exception):
    def __init__(self,message):
        print(message)
    def logout(self):
        print("logout")

class Google:

    def __init__(self,name,email,password,device):
        self.name = name
        self.email = email
        self.password = password
        self.device = device

    def login(self,email,password,device):
        if device != self.device:
            raise SecurityError("teri to lg gyi re")
        if email == self.email and password == self.password:
            print("welcome")
        else:
            print("login error")

obj = Google('vikas','vk.@gmail.com','12345','Android')

try:
    obj.login('vk.@gmail.com','12345','window')
except SecurityError as e:
    e.logout()
else:
    print(obj.name)
finally:
    print("database connection closed")
```
