### What are lists
List is a data type where we can store multiple items under i name. More technically ,list act like dynamic arrays which means we can add more items on the fly

### How lists are stored in memory?

```python
L = [1,2,3,1]
print(id(L))
print(id(L[0]))
print(id(L[1]))
print(id(L[2]))
print(id(L[3]))
print(id(1))
print(id(2))
print(id(3))
```

### Creating Lists

```python
# Empty
print([])
# 1D -->Homogeneous
print([1,2,3,4,5])
# 2D --> Heterogeneous
print([1,2,3,[4,5,6]])
# 3D --> Homogeneous
print([[[7,1],[7,8]],[[4,6],[1,2]]])
# Heterogeneous
print([1,'vikas',True,5+6j])
# Using Type conversion
print(list('vikas'))
```

### Accessing items from a List

```python
# Indexing
L = [1,2,3,[4,5]]
print(L[0])
print(L[3])
print(L[-1])
print(L[-3])
print(L[-1][-1])
print(L[-1][1])
# Slicing
L = [1,2,3,4,5,6,7,8,9]
print(L[:])
print(L[0:6])
print(L[2:8])
print(L[::2])
print(L[-5:-1:1])
print(L[-5::1])
print(L[::-1])
```

### Adding Items to a list

```python
# append-->adding one item to the end of list
L = [1,2,3,4,5,5]
L.append(True)
L.append('vikas')
L.append([1,3,6])
print(L)
```

```python
# extend --> adding multiple items
L = [1,4,3,2,5,6,7]
L.extend([12,13,14])
L.extend('vikas')
print(L)
```

```python
# insert(index,value) --.to insert desired location
L = [1,2,3,4,5]
L.insert(2,100)
L.insert(4,[200,300])
print(L)
```

### Editing items in list

```python
L = [1,2,3,4,5]
L[-1] = 500
print(L)
L[1:3] = [100,200]
print(L)
```

### Deleting items from a list

```python
# del
L = [1,2,3,4,5,6,7,8,9]
print(L)
del L[0]
print(L)
del L[1:4]
print(L)
del L
print(L)
```

```python
# Remove
L = [1,2,3,4,5,6]
L.remove(4)
print(L)
```

```python
# Pop
L = [1,2,3,4,5,6]
L.pop()
print(L)
L.pop(2)
print(L)
```

```python
# clear
L = [1,2,3,4,5,6]
print(L)
L.clear()
print(L)
```

### Operation on lists
* Arithmetic
* Membership
* Loop

```python
# arithmetic (+, *)
l1 = [1,2,3]
l2 = [4,5,6]
# Concatenation / Merge
print(l1+l2)
print(l1*3)
```

```python
# Membership
l1 = [1,2,3,4,5,6]
l2 = [4,5,6,4,[7,8,9]]
print(5 in l1)
print(5 not in l1)
print(7 in l2)
print([7,8,9] in l2)
```

```python
# loop
for i in [1,2,3,4,5,[12,23,34]]:
    print(i)
for i in [[[1,6],[8,9]],[[5,6],[2,4]]]:
    print(i)
```

### List Functions

```python
# len / min / max / sorted
L = [2,4,3,1,5,6]
print(len(L))
print(min(L))
print(max(L))
print(sorted(L,reverse=True))
```

```python
# count
L = [1,2,3,1,2,5,6,2,1]
print(L.count(1))
print(L.count(L[0]))
```

```python
# index --> tells the first occurance
L = [1,2,3,1,2,5,6,2,1]
L.index(2)
```

```python
# reverse -->permanent
L = [1,2,3,1,2,5,6]
print(L)
L.reverse()
print(L)
```

```python
# sort -->permanent(vs sorted-->not permanent)
L = [1,2,3,1,2,5,6]
print(L)
print(sorted(L))
print(L)
L.sort()
print(L)
```

```python
# copy
L = [1,2,3,1,2,5,6,2,1]
print(L)
print(id(L))
L1 = L.copy()
print(L1)
print(id(L1))
```

# List Comprehension

List Comprehension provides a concise way of creating lists

newlist  = [expression  for  item  in  iterable  if  condition ==  True]

#### Advantages of List Comprehension
* More time-efficient and space-efficient than loops
* Require fewer lines of code
* Transforms iterative statements into a formula

```python
# Add 1 to 10 number to a list
L = []
for i in range(1,11):
    L.append(i)
print(L)
```

```python
L = [i for i in range(1,11)]
print(L)
```

```python
# scalar multiplication on a vector
v = [2,3,4]
s = 3
print(v*s)
print(v)
s = -3
print(v*s)
print(v)
```

```python
v = [2,3,4]
s = -3
x = []
for i in v:
    x.append(i*s)
print(x)
```

```python
v = [2,3,4]
s = -3
[s*i for i in v]
```

```python
# Add squares
L = [1,2,3,4,5]

[i**2 for i in L]
```

```python
 # print all numbers divisible by 5 in the range of 1 to 50

[i for i in range(1,51) if i%5==0]
```

```python
# find language which start with letter p
languages = ['java','python','php','c','javascript']
# [i for i in languages if i[0]=='p']
[i for i in languages if i.startswith('p')]
```

```python
# nested if with list comprehension
basket = ['apple','guava','cherry','banana']
my_fruits = ['apple','kiwi','grapes','banana']
# add new list from my_fruits and items if the fruit exists in basket and also starts with 'a'

# [fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]

[fruit for fruit in my_fruits if fruit in basket and fruit.startswith('a')]
```

```python
# Print a (3,3) matrix using list comprehension -> Nested List comprehension
[[i*j for i in range(1,4)] for j in range(1,4)]  # moving from outside to inside
```

```python
# cartesian product -> List comprehension on 2 lists together
l1 = [1,2,3,4]
l2 = [5,6,7,8]
[[(i,j) for i in l1] for j in l2]
```

```python
l1 = [1,2,3,4]
l2 = [5,6,7,8]
[(i,j) for i in l1 for j in l2]
```

```python
l1 = [1,2,3,4]
l2 = [5,6,7,8]
[[i*j for i in l1] for j in l2]
```

```python
l1 = [1,2,3,4]
l2 = [5,6,7,8]
[i*j for i in l1 for j in l2]
```

## 2 ways to traverse a list

* itemwise
* indexwise

```python
# itemwise
L = [1,2,3,4]
for i in L:
    print(i)
```

```python
# indexwise
L = [1,2,3,4]
for i in range(len(L)):
    print(i,L[i])
```

### Zip()
the zip() function returns a zip object , which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together
If the passed iterators have different lengths , the iterator with the least items decides the length of the new iterator

```python
# write a program to add item of 2 lists indexwise

l1 = [1,2,3,4]
l2 = [-1,-2,-3,-4]
print(list(zip(l1,l2)))
[i+j for i,j in zip(l1,l2)]
```

```python
list1 = [2,4,6,10,1]
n = len(list1)
L = []
for i in range(0,n):
    j = 0
    s = 0
    while j<n :
        if list1[j]>=list1[i]:
            s += list1[j]
        j += 1
    L.append(s)
print(L)
```

```python
num1 = [23,45,67,78,89,34]
num2 = [34,89,55,56,39,67]
L = []
for i in range(len(num1)):
    if num1[i] in num2:
        L.append(num1[i])
L.sort()
print(L)
```

```python
l = ['1ac21','23fg','456','098d','1','kls']
L = []
for i in range(len(l)):
    prod = 1
    for j in l[i]:
        if j.isdigit():
            prod *= int(j)
    L.append(prod)
print(L)
```
