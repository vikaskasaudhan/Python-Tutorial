```python
# using context manager (with)
with open("sample.txt",'w') as f:
    f.write("hello Vikas Bhai")
```

```python
# try f.read()
with open("sample.txt",'r') as f:
    print(f.read())
```

```python
# moving within a file -> 10 char then 10 char
with open('sample3.txt','r') as f:
    print(f.read(10))
    print(f.read(10))
```

```python
# benefits to load a big file in memory
big_L = [' I Love You' for i in range(1000)]

with open('big.txt','w') as f:
    f.writelines(big_L)
```

```python
with open('big.txt','r') as f:

    chunk_size = 20

    while len(f.read(chunk_size))>0:
        print(f.read(chunk_size),end='***')
        f.read(chunk_size)
```

```python
# seek and tell function
with open('sample3.txt','r') as f:
    print(f.read(10))
    print(f.tell())
    f.seek(0)
    print(f.read(10))
    print(f.tell())
    f.seek(15)
    print(f.read(10))
    print(f.tell())
```

```python
# seek during write
with open('sample.txt','w') as f:
    f.write("Vikas Kasaudhan")
    f.seek(0)
    f.write('X')
```

```python
# working with binary file
with open('vikas.jpeg','rb') as f:
    with open('vikas_copy.jpeg','wb') as wb:
        wb.write(f.read())
```

## Serialization and Deserialization

```python
# Serialisation
# serialization and deserialization dictionary
import json
L = [1,2,3,4,5,6]
with open('demo.json','w') as f:
    json.dump(L,f)
```

```python
d = {
    'name':'vikas',
    'age':21,
    'gender':'male'
}
with open('demo.json','w') as f:
    json.dump(d,f)
```

```python
# deserialisation
import json

with open('demo.json','r') as f:
    d = json.load(f)
    print(d)
    print(type(d))
```

```python
# serialization and deserialization tuple
t = (1,2,3,4,5)
with open('demo.json','w') as f:
    json.dump(t,f)
```

```python
with open('demo.json','r') as f:
    t = json.load(f)
    print(t)
    print(type(t))
```

### serialization and deserialization of custom object

```python
class Person:

    def __init__(self,fname,lname,age,gender):
        self.fname = fname
        self.lname = lname
        self.age = age
        self.gender = gender

#format to printed in 
# -> Vikas Kasaudhan age - > 21 gender -> male
```

```python
person = Person('Vikas',"Kasaudhan",21,'male')
```

```python
# as a string

import json

def show_object(person):
    if isinstance(person,Person):
        return "{} {} age -> {} gender -> {}".format(person.fname,person.lname,person.age,person.gender)

with open('demo.json','w') as f:
    json.dump(person,f,default=show_object)
```

```python
import json

with open('demo.json','r') as f:
    d = json.load(f)
    print(d)
    print(type(d))
```

## Pickling and Unpickling

```python
class Person:

    def __init__(self,name,age):
        self.name = name
        self.age = age

    def display_info(self):
        print('hi my name is',self.name,'and I am',self.age,'years old')
```

```python
p = Person('vikas kasaudhan',21)
```

```python
#pickle dumo
import pickle
with open('person.pkl','wb') as f:
    pickle.dump(p,f)
```

```python
# pickle load

with open('person.pkl','rb') as f:
    p = pickle.load(f)

p.display_info()
```
