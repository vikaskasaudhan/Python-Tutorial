# Measuring Time

```python
# not used in industry because it depends on machine 
import time

start = time.time()
for i in range(1,101):
    print(i,end=' ')
end = time.time()
print(end-start)
```

```python
start = time.time()
i = 1
while i<=100:
    print(i,end=' ')
    i += 1
end = time.time()
print(end-start)
```

### Problem with this approach
1. Different time for different algorithm (doing good)
2. Time varies if implementation changes
3. Different machines different time
4. Does not work for extremely small input
5. Time varies for different inputs,but can not establish a relationship
