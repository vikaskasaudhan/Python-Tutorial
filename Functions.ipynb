### *args  and  **kwargs

*args  and  **kwargs  are special Python keywords that are used to pass the variable length of arguments to a function

```python
# *args
# allows us to pass a variable number of non-keyword arguments to a function

def multiply(*args):
    prod = 1
    for i in args:
        prod *= i
    return prod
```

```python
multiply(1,2,3,4,5,6,7,8,9)
```

```python
# **kwargs
# **kwargs allows us to pass any number of keywords arguments.
# Keyword arguments mean that they contain a key-value pair, like a Python dictionary.

def display(**kwargs):

    for (key,value) in kwargs.items():
        print(key,"->",value)
```

```python
display(India='New Delhi',Nepal="kathmandu",srilanka="colombo",pakistan="Islamabad")
```

### Points to remember while using  *args  and  **kwargs
* order of the arguments matter (normal -> *args -> **kwargs)
* The words  *args  and  **kwargs  are only a convention , we can use any name of our choice

## Lambda Function

A lambda function is a small anonymous function.

A lambda function can take any number of arguments, but can only  have one expression.

```python
# x -> x^2
a = lambda x:x**2
a(2)
```

```python
# x , y -> x+y
a = lambda x,y:x+y
a(3,6)
```

### Difference between lambda vs Normal Function
* No name
* lambda has no return value (infact, return a function)
* lambda is written in 1 line
* not reusable

#### Then why use lambda function?
They are used with HOF(higher order function)

```python
# check if a string has 'a'
a = lambda s:'a' in s
a('vikas')
```

```python
# odd or even
a = lambda x:'even' if x%2==0 else 'odd'
a(5)
```

```python
# example

def square(x):
    return x**2

# HOF
def transform(f,L):
    output = []
    for i in L:
        output.append(f(i))
    print(output)
L = [1,2,3,4,5]
transform(square,L)
```

```python
# HOF
def transform(f,L):
    output = []
    for i in L:
        output.append(f(i))
    print(output)
L = [1,2,3,4,5]
transform(lambda x:x**2,L)
```

```python
def closest_distance(L,Q):
    dis = ((L[0][0]-Q[0])**2 + (L[0][1]-Q[1])**2)**0.5
    t = L[0]
    for i in range(1,len(L)):
        dis2 = ((L[i][0]-Q[0])**2 + (L[i][1]-Q[1])**2)**0.5
        if dis2<dis:
            dis = dis2
            t = L[i]
    return t

L = [(7,7),(2,2),(3,3),(4,4),(5,5),(1,2)]
Q = (1,2)
point = closest_distance(L,Q)
print("nearest to ",Q, "is",point)
```

```python
def most_occurence_word(s):
    L = []
    for i in s.split():
        t = (i,s.count(i))
        L.append(t)
    t1 = L[0]
    for i in range(1,len(L)):
        if L[i][1]>t1[1]:
            t1 = L[i]
    return t1
s = input("enter string")
occur = most_occurence_word(s)
print(occur[0],"->",occur[1])
```

```python
def new_dict(L):
    dic = {}
    for i in L:
        for (key,value) in i.items():
            dic[key] = value
    return dic

dict1 = {1:10,2:20}
dict2 = {3:30,4:40}
dict3 = {5:50,6:60}
L = [dict1,dict2,dict3]
new_dic = new_dict(L)
print(new_dic)
```

```python
def unique_list(L):
    l = []
    for i in L:
        if i not in l:
            l.append(i)
    return l
L = [1,1,2,2,2,3,3,3,4,4,4,5,5,5]
new_list = unique_list(L)
print(new_list)
```

```python
def sequence(s):
    l = []
    for i in s.split(sep='-'):
        l.append(i)
    l = sorted(l)
    return l

s = 'green-red-yellow-black-white'
sorted_sequence = "-".join(sequence(s))
print(sorted_sequence)
```

```python
def upper_lower(s):
    count1 = 0
    count2 = 0
    for i in range(len(s)):
        if s[i].islower():
            count1 += 1
        if s[i].isupper():
            count2 += 1
    l = []
    l.append(count1)
    l.append(count2)
    return l
s = input("enter string")
l = upper_lower(s)
print("no of lowercase letters = ",l[0])
print("no of uppercase letters = ",l[1])
```

```python
def print_even(L):
    l = []
    for i in L:
        if i%2==0:
            l.append(i)
    return l
L = [1,2,3,4,5,6,7,8,9]
l = print_even(L)
print(l)
```

```python
def perfect(num):
    sum = 0
    for i in range(1,num):
        if num%i==0:
            sum += i
    if sum==num:
        return 'perfect'
    else:
        return 'not perfect'
n = int(input("enter number"))
print(perfect(n))
```
